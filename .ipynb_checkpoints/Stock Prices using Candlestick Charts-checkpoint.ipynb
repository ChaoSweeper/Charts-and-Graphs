# Stock price tracker w/ candlestick charts

## 1. Check the stock price history for a period of time

### 1.1 Import the required libraries 

```python
import matplotlib.pyplot as plt
import pandas_datareader as web
import mplfinance as mpf
import datetime as dt
```

### 1.2 Set the start and end date to collect the data from

```python
start_time = dt.datetime(2019, 1, 1)
end_time = dt.datetime(2020, 1, 1)
```

### 1.3 Pull the data from yahoos financial website

```python
data = web.DataReader("LMT", "yahoo", start_time, end_time)
data
```

### 1.4 Display a basic chart over the whole time period

```python
# print(mpf.available_styles())
mpf.plot(data, type="candle", style="nightclouds")
```

### 1.5 Set the colors and style for the candlestick charts

```python
colors = mpf.make_marketcolors(
    up="#00ff00", down="#ff0000", wick="inherit", edge="inherit", volume="in"
)
```

### 1.6 Display the data

```python
mpf_style = mpf.make_mpf_style(base_mpf_style="nightclouds", marketcolors=colors)
mpf.plot(data, type="candle", style=mpf_style, volume=True)
```

## 2. Closer look at when prices dip between March 15 - May 28

### 2.1 Set the start and end date to display a certain time period

```python
start_time = dt.datetime(2019, 3, 15)
end_time = dt.datetime(2019, 5, 28)
```

### 2.2 Display the the raw data

```python
data = web.DataReader("LMT", "yahoo", start_time, end_time)
data
```

### 2.3 Display the data as a candlestick chart

```python
mpf.plot(data, type="candle", style="mike")
```

### 2.4 Final look w/ price volmue

```python
mpf_style = mpf.make_mpf_style(base_mpf_style="mike")
mpf.plot(data, type="candle", style=mpf_style, volume=True)
```
